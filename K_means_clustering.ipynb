```python
import pandas as pd
import numpy as np 
import matplotlib.pylab as plt
import seaborn as sns
from sklearn.cluster import KMeans              #For clustering
from scipy.spatial.distance import cdist        #For calculating within sum of squares distances
from sklearn.preprocessing import MinMaxScaler
import warnings
warnings.filterwarnings("ignore")
```

```python
# Creating Random 1000 variables X and Y
```

```python
X = np.random.uniform(0,1,1000)                 # This implies I want to create 1000 values randomly between 0 and 1
X
```

```python
Y = np.random.uniform(0,1,1000)                   # This implies I want to create 1000 values randomly between 0 and 1
Y
```

```python
df_xy = pd.DataFrame(columns=["X","Y"])
df_xy.X = X                                     # This will paste all the created random values below X column
df_xy.Y = Y                                     # This will paste all the created random values below Y column
df_xy.plot(x ="X", y="Y", kind = "scatter")
```

```python
# Creating the cluster using kmeans algorithm
```

```python
model1 = KMeans(n_clusters=5).fit(df_xy)
df_xy.plot(x="X",y="Y",c = model1.labels_, kind = "scatter",s=15, cmap = plt.cm.coolwarm)     # .labels_ will create a cluster on the basis of model1, s=15 means size of the data points in a cluster will be 15, it can be anything, plt.cm.coolwarm will create a colored cluster it is completely optional 
```

```python
# Performing k_means clustering on University data set
```

```python
Univ = pd.read_csv("G:/data sceince/Python/Clustering/Hierarchical clustering/Universities.csv")
```

```python
# First we we scale this data as the exceptance column is having very high values and therefore we are getting high distances so we will normalize this data first
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(Univ.iloc[:,1:])        # I will apply the user defined function norm_func on Univ data set for all the rows and for specified coulmns
df_norm
```

```python
# Creating the ELBOW curve for deciding the appropriate number of cluster
```

```python
wcss = []                                         # creating an empty list for storing the within sum of squares distances
for i in range(1,11):                        
    kmeans = KMeans(n_clusters = i,random_state=0)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title("ELBOW CURVE")
plt.xlabel("Number of clusters in K")
plt.ylabel("WCSS")
    
```

```python
# Building a final model using KMeans
```

```python
model = KMeans(n_clusters = 4)                      # From the above scree plot we come to know that we can consider K =4
model.fit(df_norm)                                  # we will train our model
model.labels_                                       # This will give cluster membership and we will understand which university belongs to which cluster
```

```python
# Converting this array to series format

md = pd.Series(model.labels_)
Univ['Clust'] = md                        # We will assign the name as Clust to the cluster membership labels
```
